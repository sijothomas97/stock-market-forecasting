```python
# Calculate a longer-term SMA (e.g., 30-day SMA)
monthly_hdfc_data['SMA5'] = monthly_hdfc_data['volume'].rolling(window=5).mean()
monthly_hdfc_data['SMA10'] = monthly_hdfc_data['volume'].rolling(window=10).mean()
monthly_hdfc_data['SMA15'] = monthly_hdfc_data['volume'].rolling(window=15).mean()
monthly_hdfc_data['SMA20'] = monthly_hdfc_data['volume'].rolling(window=20).mean()
monthly_hdfc_data['SMA30'] = monthly_hdfc_data['volume'].rolling(window=30).mean()

fig, axs = plt.subplots(5, sharex=True, sharey=True, figsize=(10, 10))

fig.suptitle('Volume of stocks')
axs[0].set_title("HDFC Bank")
axs[0].bar(monthly_hdfc_data.index, monthly_hdfc_data['volume'], width=15)
axs[1].set_title("Reliance Industries")
axs[1].bar(monthly_reliance_data.index, monthly_reliance_data['volume'], width=15)
axs[2].set_title("Sun Pharmaceutical Industries")
axs[2].bar(monthly_sunpharma_data.index, monthly_sunpharma_data['volume'], width=15)
axs[3].set_title("Tata Steel")

# Show Plot
plt.show()

# fig.show()
# Create a plot to visualize volume trends
plt.figure(figsize=(12, 6))
plt.plot(monthly_hdfc_data.index, monthly_hdfc_data['volume'], label='Volume', color='black', alpha=0.5)
plt.plot(monthly_hdfc_data.index, monthly_hdfc_data['SMA5'], label='5-day SMA', color='blue')
plt.plot(monthly_hdfc_data.index, monthly_hdfc_data['SMA10'], label='10-day SMA', color='orange')
plt.plot(monthly_hdfc_data.index, monthly_hdfc_data['SMA15'], label='15-day SMA', color='green')
plt.plot(monthly_hdfc_data.index, monthly_hdfc_data['SMA20'], label='20-day SMA', color='red')
plt.plot(monthly_hdfc_data.index, monthly_hdfc_data['SMA30'], label='30-day SMA', color='yellow')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Trading Volume and Moving Averages')
plt.legend()
plt.grid(True)
plt.show()
```

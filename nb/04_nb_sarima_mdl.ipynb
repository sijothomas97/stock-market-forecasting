```python
import itertools
# import warningsz
import statsmodels.api as sm
# from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima_model import ARIMA
import numpy as np
```

```python
# Model Trainin and Fitting
y_to_train = train_hdfc['first_difference'] # dataset to train
y_to_test = test_hdfc['first_difference'] # last X months for test
y_to_val = y_to_test

# warnings.filterwarnings("ignore") # specify to ignore warning messages

p = d = q = range(0, 2)
seasonal_period = 12
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2],seasonal_period) for x in list(itertools.product(p, d, q))]


for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(y_to_train,
                                      order=param,
                                      seasonal_order=param_seasonal,
                                      enforce_invertibility=False)
            results_hdfc = mod.fit()
        except Exception as ex:
            continue

order = (0, 0, 1) #Using Values from the previous step
seasonal_order = (1, 1, 1, 12)
model = sm.tsa.statespace.SARIMAX(y_to_train,
                            order=order,
                            seasonal_order=seasonal_order,
                            enforce_invertibility=False)
results_hdfc = model.fit()
print(results_hdfc.summary())
```
